```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

print("Biblioteki do wizualizacji danych zaimportowane.")
```

```python
#Dane
x = [1,2,3,4,5]
y = [2,3,5,7,11]

#Tworzenie wykresu
plt.plot(x, y)

#Dodanie tytułu i etykiet osi
plt.title("Mój Pierwszy Wykres Liniowy")
plt.xlabel("Oś X")
plt.ylabel("Oś Y")

#Wyświetlenie wykresu
plt.show()

x = np.arange(0,10,0.5) #Liczby od 0 do 9.5 z krokiem 0.5
y1 = x * 2 
y2 = x ** 2

plt.figure(figsize=(8,5)) #Ustawienie rozmiaru wykresu (szerokość, wysokość w calach)

plt.plot(x, y1, color='blue', linestyle='--', marker='o', label='Funkcja liniowa (2x)')
plt.plot(x, y2, color='red', linestyle='-', marker='x', label="Funkcja kwadratowa (x^2)")

plt.title("Wykres Liniowy z Wieloma Seriami")
plt.xlabel("Wartość X")
plt.ylabel("Wartość Y")
plt.grid(True) #Dodanie siatki
plt.legend() #Dodanie legendy (pokazuje etykiety z 'label')

plt.show()
```

```python
#Generowanie przykładowych danych
np.random.seed(42)
punkty_x = np.random.rand(50)*100
punkty_y = punkty_x + np.random.randn(50) * 20 #Dodajemy trochę szumu

plt.figure(figsize=(7,6))
plt.scatter(punkty_x, punkty_y, color='green', alpha=0.6, s=50) # alpha to przezroczystość, s to rozmiar punktu

plt.title("Wykres Punktowy: Korelacja Wiek vs Dochód")
plt.xlabel("Wiek (lata)")
plt.ylabel("Dochód (tys. zł)")
plt.grid(True)

plt.show()
```

```python
# Generowanie przykładowych danych - rozkład normalny
dane_histogram = np.random.randn(1000) * 15 + 170 #Średnia 170, odchylenie 15

plt.figure(figsize=(8,5))
plt.hist(dane_histogram, bins=30, color='skyblue', edgecolor='black', alpha=0.7) #bins - liczba koszy

plt.title("Histogram: Rozkład Wzrostu Populacji")
plt.xlabel("Wzrost (cm)")
plt.ylabel("Częstość")
plt.grid(axis='y', alpha=0.75) #siatka tylko na osi Y

plt.show
```

```python
miesiace = np.array(['Styczeń', 'Luty', 'Marzec', 'Kwiecień', 'Maj'])
sprzedaz = np.array([100, 120, 150, 130, 180])

plt.plot(miesiace, sprzedaz, color='orange', linestyle=':')

plt.title("Trend Sprzedaży 2025")
plt.xlabel("Miesiąc")
plt.ylabel("Sprzedaż (szt.)")

```

```python
ocena_testu = np.array([60, 75 ,80, 92, 65, 70, 88, 95, 78, 85])
godziny_nauki = np.array([2, 3, 4, 5, 2.5, 3.5, 4.8, 5.5, 4.1, 4.9])

plt.scatter(ocena_testu, godziny_nauki, s=70, color='purple')
plt.title("Zależność między Godzinami Nauki, a Oceną Testu")
plt.xlabel("Ocena Testu (%)")
plt.ylabel("Godziny Nauki")
```

```python
dane_histogram = np.random.normal(loc=50, scale=10, size=500) #loc to średnia, scale to odchylenie standardowe
plt.hist(dane_histogram, bins=20, color='lightgreen', edgecolor='black')
plt.title("Rozkład Wyników Egzaminu")
plt.xlabel("Wynik Egzaminu")
plt.ylabel("Liczba Studentów")          
```

```python
#dane z poprzedniego histogramu
dane_histogram = np.random.normal(loc=50, scale=10, size=500)

plt.figure(figsize=(8, 5))
sns.histplot(dane_histogram, bins=20, kde=True, color='purple') #kde=True dodaje estymację gęstości jądrowej
plt.title("Histogram z Seaborn: Rozkład Wyników Egzaminu")
plt.xlabel("Wynik Egzaminu")
plt.ylabel("Liczba Studentów")
plt.show()
```

```python
data_boxplot = pd.DataFrame({'Kategoria':['A'] * 30 + ['B'] * 40 + ['C'] * 30,
                             'Wartość': np.concatenate([np.random.normal(loc=10, scale=2, size=30),
                                                        np.random.normal(loc=15, scale=3, size=40),
                                                        np.random.normal(loc=8, scale=1.5, size=30)
                                                       ])
                            })

plt.figure(figsize=(8, 6))
sns.boxplot(x='Kategoria', y='Wartość', data=data_boxplot, palette='viridis') #palette zmienia schemat kolorów

plt.title("Wykres Pudełkowy Warości dla Różnych Kategorii")
plt.xlabel("Kategoria")
plt.ylabel("Wartość")
plt.show()
```
